```python
"""
Docstring:
    A) Introduction
        This aim of this game is to help little Missy, a young girl scout,
        escape a haunted forest after she got lost. It consist of three main rounds.
    B) Bugs and errors;
        None.
"""

from sys import exit

#start of the game

def start():
    print("""Hi, my name is Missy! I am so glad I found you.\n I came for a trek with my friends but got lost.
    It is such a scary and big forest.\n Could you please help me find my way out.""")
    
    input("< press enter to continue>\n")
    starting_path()

# round 1
def starting_path():
    print("""You are in a forest, there is quicksand to the west and a path leads to the east.\n 
    Which path do you take?""")
    
    #nested conditions.
    
    decision = input("<east or west> \n")
    
    if "east" in decision or "EAST" in decision or "East" in decision:
        print("you are off to the path")
        input("<press enter to continue>\n")
        path_area()
    
    elif "west" in decision or "WEST" in decision or "West" in decision:
        print("""You walk for one minute but eventually get stuck in the sand.
        You and Missy stay in the forest, stuck for days and finally die of hunger and dehydration.""")
        
        print("Do you want to make another choice?")
        
        answer = input("<type 'yes or no'>\n")
           
        if "yes" in answer or "Yes" in answer or "YES" in answer:
            input("<press enter to continue\n")
            starting_path()
            
        elif "no" in answer or "No" in answer or "NO" in answer:
            input("<press enter to continue>\n")
            fail()
            
        else:
            print("""please check your answer again. Type in 'yes or no' only.""")
            input("<press enter to continue>\n")
            
    else:
        print("Please put in 'east or 'west'")
        input("<press enter to continue>\n")
        starting_path()

#round 2

def path_area():
    print(""" It is cold, dark and scary. You can hear owls hooting from a distance,\n
    the crackling of leaves beneath your feet makes little Missy terrified.\n
    Suddenly, Missy screams! A huge green ghost appears infront of you.\n
    Little Missy clutches on your leg petrified.""")
    
    print(""" Green ghost: I am the ghots of halloween eve!
    Who dares cross my path!""")
    Heroes_name = input("My... My.... My na... My name is..")
    
    
    print(f"\n Green ghost: Aha! {Heroes_name} why do you dare cross my path?")
    
    print(f"""{Heroes_name}: I... I just want to help little missy get her way out of the forest.\n
    We didn't mean any harm to you. Please let us through.\n
    Green ghost: I'll let you pass but only if you can answer this simple riddle.
    You have only 3 guesses.""")
    
    print("""It has a long neck, A name of a bird, Feeds on cargo of ships, 
    It's not alive,
    
    What is it?""")
    
    #while loops
    
    guesses = 3
    
    while guesses > 0:
    
        choice = input()
    
        if "crane" in choice or "a crane" in choice:
            print(" Brave soul! continue on your path and get little Missy out of here")
            tunnel()
    
        elif "crane" not in choice:
            print("Wrong answer, try again?")
            print("<Hint; it rhymes with PANE>")
            guesses -= 1
          
              
        else:
            print("Ooops, code error!")
     
    if guesses == 0:
        print("The answer was CRANE!")
        fail()
                    
 #round 3   
            
def tunnel():
    print("""You and missy continue down the path, you see a big tunnel.\n
    Missy shouts happily, this is the way out! I am sure of it! However, a huge
    wooden door barricades the entrance. You look around frantically searching 
    for a way to open the door.
    \n You collect the following; an axe and matchbox. 
    What do you use to break the wooden door""")
    
    tool = input("\n< axe or matchbox >")
    
    #conditional statement
    
    if "axe" in tool or "Axe" in tool or "AXE" in tool:
        
        print("""You successfully break down the door. You walk down the tunnel
        with little Missy in your hand. You successfully lead missy out of the
        haunted forest! You are a superstar!!""")
        exit(0)
    
    elif "matchbox" in tool or "Matchbox" in tool or "MATCHBOX" in tool:
    
        print("""You start an uncontrollable fire. You burn down the forest.
        You, little Missy and everything in the forest suffocate and burn to 
        death.""")
        fail()
    
    else:
        print("please choose between axe or matchbox.")
        
#fail function

def fail():
    print(f"""You lost! Better luck next time.""")
    exit(0)
        
                 
    
    
    
    
###############################################################################
#Game Start
###############################################################################
start()

    
```
